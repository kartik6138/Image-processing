```python
#Image Sharpen
from PIL import Image
from PIL import ImageFilter
import matplotlib.pyplot as plt

#load the image
my_img=Image.open('22.jpg')

#use sharpen function
sharp=my_img.filter(ImageFilter.SHARPEN)

#save the image
sharp.save("D:/sharpen_image.jpg")
sharp.show()
plt.imshow(sharp)
plt.show()
```

```python
#Image flipp
import matplotlib.pyplot as plt

#load the image
img=Image.open('22.jpg')
plt.imshow(img)
plt.show()

#use the flip function
flip=img.transpose(Image.FLIP_LEFT_RIGHT)

#save the image
flip.save("D:/flip_img.jpg")
plt.imshow(flip)
plt.show
```
